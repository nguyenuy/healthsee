```python
from bokeh.sampledata import us_states
from bokeh.plotting import *
us_states = us_states.data.copy()
del us_states["HI"]
del us_states["AK"]

# separate latitude and longitude points for the borders
#   of the states.
state_xs = [us_states[code]["lons"] for code in us_states]
state_ys = [us_states[code]["lats"] for code in us_states]

def get_us_figure(title):
    # init figure
    p = figure(title=title, 
               toolbar_location="left", plot_width=1100, plot_height=700)

    # Draw state lines
    p.patches(state_xs, state_ys, fill_alpha=0.0,
        line_color="#884444", line_width=1.5)
    
    return p
```

```python
# Plot of Similar Zipcodes
zipcode = 30312
p = get_us_figure("Zipcodes similar to Zipcode: " + str(zipcode))

similar_zipcodes = hd.get_similar_zipcodes(zipcode)
x = [hd.get_zipcode_coordinates(z)[1] for z in similar_zipcodes]
y = [hd.get_zipcode_coordinates(z)[0] for z in similar_zipcodes]

zip_coords = hd.get_zipcode_coordinates(zipcode)


# The scatter markers
p.circle(x, y, size=8, color='navy', alpha=1)
p.circle(zip_coords[1], zip_coords[0], size=8, color='red', alpha=1)

show(p)
```
